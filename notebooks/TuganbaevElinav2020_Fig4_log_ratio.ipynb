```python
import pandas as pd 
```

```python
feature_table = pd.read_csv("../data/Tuganbaev_Elinav/feature_table-deblur-tuganbaev.txt", sep='\t', skiprows = 1, index_col=0)
feature_table.head()
```

```python
taxonomy_table = pd.read_csv('../data/Tuganbaev_Elinav/Tb_taxonomy.tsv', sep='\t', index_col=0)
taxonomy_table.head()
```

```python
taxonomy_table[['Taxon']]
```

```python
feature_table_with_taxon = pd.merge(feature_table, taxonomy_table[['Taxon']], right_index = True, left_index =True)
feature_table_with_taxon.head()
```

```python
feature_table_with_taxon.set_index('Taxon', inplace=True)
feature_table_with_taxon
```

```python
feature_table_with_taxon_flipped = feature_table_with_taxon.T
feature_table_with_taxon_flipped.head()
```

```python
lum_muc_metadata = pd.read_csv('../data/Tuganbaev_Elinav/Tb_metadata_lum_muc.txt', sep ='\t', index_col=0)
lum_muc_metadata.head()
```

```python
dictionary = {0:24, 4:4, 12:12, 16:16, 20:20}
lum_muc_metadata = lum_muc_metadata.replace({'zt':dictionary})
lum_muc_metadata.head()
```

```python
feature_table_plus_lum_muc_merged = pd.merge(feature_table_with_taxon_flipped, 
                                              lum_muc_metadata[['zt', 'cage', 'content_source']], left_index = True, right_index = True)
feature_table_plus_lum_muc_merged.head(10)
```

```python
lacto_list = [col for col in feature_table_plus_lum_muc_merged.columns if 'Lactobacillus' in col]
lacto_list
```

```python
feature_table_plus_lum_muc_merged['Lactobacillus'] = feature_table_plus_lum_muc_merged[[col for col in feature_table_plus_lum_muc_merged.columns if 'Lactobacillus' in col]].sum(axis="columns")
feature_table_plus_lum_muc_merged.head(8)
```

```python
feature_table_plus_lum_muc_merged['Allobaculum'] = feature_table_plus_lum_muc_merged[[col for col in feature_table_plus_lum_muc_merged.columns if 'Allobaculum' in col]].sum(axis="columns")
feature_table_plus_lum_muc_merged.head(8)
```

```python
feature_table_plus_lum_muc_merged['Bifidobacterium'] = feature_table_plus_lum_muc_merged[[col for col in feature_table_plus_lum_muc_merged.columns if 'Bifidobacterium' in col]].sum(axis="columns")
feature_table_plus_lum_muc_merged.head(8)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statannot import add_stat_annotation
%matplotlib inline 
```

```python
feature_table_plus_lum_muc_merged['log_ratios_L_to_A'] = np.log(feature_table_plus_lum_muc_merged['Lactobacillus']) - np.log(feature_table_plus_lum_muc_merged['Allobaculum'])
feature_table_plus_lum_muc_merged.head()
```

```python
#Lactobacillus vs Allobaculum
ax = sns.boxplot(data=feature_table_plus_lum_muc_merged, x='zt', y='log_ratios_L_to_A', hue ='content_source')
#both were in the top 5% of DEICODE importance
```

```python
box_pairs_input = [((zt, 'Luminal'), (zt, 'Mucosal')) for zt in feature_table_plus_lum_muc_merged['zt'].unique()]
box_pairs_input
```

```python
#check to see if lum vs muc is significantly different
ax = sns.boxplot(data=feature_table_plus_lum_muc_merged, x='zt', y='log_ratios_L_to_A', hue ='content_source')
ax = sns.swarmplot(data=feature_table_plus_lum_muc_merged, x='zt', y='log_ratios_L_to_A', hue ='content_source', 
                   edgecolor='k', linewidth=2, size=10, dodge=True)
add_stat_annotation(ax, data=feature_table_plus_lum_muc_merged, x='zt', y='log_ratios_L_to_A', hue ='content_source',
                                       box_pairs= box_pairs_input,test='Mann-Whitney', text_format='star', loc='outside', verbose=2)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.savefig('../figures/log_ratio_Lacto_to_Allo', dpi=300, bbox_inches = 'tight')
```

```python
Bifido_plus_lacto['log_ratio_Lacto_to_Bifido'] = np.log(Bifido_plus_lacto['k__Bacteria; p__Firmicutes; c__Bacilli; o__Lactobacillales; f__Lactobacillaceae; g__Lactobacillus'])- np.log(Bifido_plus_lacto['k__Bacteria; p__Actinobacteria; c__Actinobacteria; o__Bifidobacteriales; f__Bifidobacteriaceae; g__Bifidobacterium'])
Bifido_plus_lacto.head()
```

```python
B_L_drop = ['F4', 'F6']
Bifido_plus_lacto_cleaned = Bifido_plus_lacto.drop(index= B_L_drop)
Bifido_plus_lacto_cleaned.shape
#this checks out
```

```python
ax = sns.boxplot(data=Bifido_plus_lacto_cleaned, x='zt', y='log_ratio_Lacto_to_Bifido', hue ='content_source')
#both were in the top 5% of DEICODE importance
```

```python
from itertools import combinations
from scipy.stats import mannwhitneyu
from scipy.stats import wilcoxon
```

```python
for key in list(combinations(Bifido_plus_lacto_cleaned.zt.unique(), 2)):
    x = Bifido_plus_lacto_cleaned[Bifido_plus_lacto_cleaned["zt"] == key[0]].log_ratio_Lacto_to_Bifido.values
    y = Bifido_plus_lacto_cleaned[Bifido_plus_lacto_cleaned["zt"] == key[1]].log_ratio_Lacto_to_Bifido.values
    if len(x) == len(y):
        w, p = wilcoxon(x, y)
        print("Combination", key)
        print(p, w)
        print("____________________")
    else:
        print("not paired")
        print("____________________")
```

```python
for key in list(combinations(Bifido_plus_lacto_cleaned.zt.unique(), 2)):
    x = Bifido_plus_lacto_cleaned[Bifido_plus_lacto_cleaned["zt"] == key[0]].log_ratio_Lacto_to_Bifido.values
    y = Bifido_plus_lacto_cleaned[Bifido_plus_lacto_cleaned["zt"] == key[1]].log_ratio_Lacto_to_Bifido.values
    w, p = mannwhitneyu(x, y)
    print("Combination", key)
    print(p, w)
    print("____________________")
```
