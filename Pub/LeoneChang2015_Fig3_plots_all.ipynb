## fecal (24h)

```python
md = pd.read_csv('../data/leone_chang_2015/leone_combined_metadata_clean2.txt',delimiter = "\t",
                 usecols=['#SampleID', 'group_diet_ZT'], index_col=0)
md.head()
```

```python
F24wUF_dm = pd.read_csv('../data/leone_chang_2015/wUF_dm_fecal.tsv', sep='\t', index_col=0)
F24wUF_dm.head()
```

```python
F24wUF_dm_md = pd.merge(md, F24wUF_dm, left_index=True, right_index=True)
F24wUF_dm_md.head()
```

```python
F24wUF_dm_md2 = F24wUF_dm_md.set_index('group_diet_ZT')
F24wUF_dm_md2.head()
```

```python
F24wUF_dm_mdID = md[['group_diet_ZT']].copy()
F24wUF_dm_mdID.head()
```

```python
F24wUF_dm_md3 = pd.merge(F24wUF_dm_md2.T, F24wUF_dm_mdID, left_index=True, right_index=True)
F24wUF_dm_md4 = F24wUF_dm_md3.set_index('group_diet_ZT')

#all fecal 24h, RC vs. HF
F24wUF_dm_md5 = F24wUF_dm_md4.filter(regex='HF')
F24wUF_dm_md6 = F24wUF_dm_md5.T.filter(regex='RC')
F24wUF_dm_md6
```

```python
F24wUF_dm_md6.shape
```

```python
wufdm = F24wUF_dm_md6.copy()
wufdm2 = wufdm.stack().reset_index()
wufdm2.columns = ['fecal_HF','fecal_RC','wUF_distance']
wufdm2.head()
```

```python
wufdm2['timepoints'] = wufdm2['fecal_HF'].str.split('_').str[2] + "_" + wufdm2['fecal_RC'].str.split('_').str[2]
wufdm2.timepoints.unique()
```

```python
zt_tps = ['2_2','6_6','10_10','14_14','18_18','22_22']
wufdm3 = wufdm2.loc[wufdm2['timepoints'].isin(zt_tps)]
wufdm4 = wufdm3.copy()
wufdm4['ZT'] = wufdm4['timepoints'].str.split('_').str[-1]
wufdm4['Controls'] = wufdm4['fecal_RC'].str.split('_').str[0]
wufdm4['Experimentals'] = wufdm4['fecal_HF'].str.split('_').str[0]
wufdm4.head()
```

```python
wufdm4.wUF_distance.mean()
```

```python
sns.set_style("white")
ax = sns.boxplot(x="ZT", y="wUF_distance", palette=sns.dark_palette("gold", n_colors=7, reverse=True), 
                 data=wufdm4, order=['2', '6', '10','14','18', '22'])
ax = sns.swarmplot(x="ZT", y="wUF_distance", palette=sns.dark_palette("gold", n_colors=7, reverse=True), 
                   edgecolor='k', linewidth=1, size=8,
                   data=wufdm4, order=['2', '6', '10','14','18', '22'])
plt.ylabel('Weighted Unifrac Between-Group Distances', fontsize=12)
plt.title('RC vs. HF (fecal 24h)', fontsize=14)
ax.axhline(0.36559547967242134, color='k', ls='--')
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['svg.fonttype'] = 'none'
#plt.savefig('../figures/fecal24hRCvsHF_boxplot_reordered.svg', dpi=300, bbox_inches="tight")
```

```python
from itertools import combinations
from scipy.stats import mannwhitneyu
from scipy.stats import wilcoxon
```

```python
for key in list(combinations(wufdm4.ZT.unique(), 2)):
    x = wufdm4[wufdm4["ZT"] == key[0]].wUF_distance.values
    y = wufdm4[wufdm4["ZT"] == key[1]].wUF_distance.values
    if len(x) == len(y):
        w, p = mannwhitneyu(x, y)
        print("Combination", key)
        print(p, w)
        print("____________________")
#
```

```python
for key in list(combinations(wufdm4.ZT.unique(), 2)):
    x = wufdm4[wufdm4["ZT"] == key[0]].wUF_distance.values
    y = wufdm4[wufdm4["ZT"] == key[1]].wUF_distance.values
    if len(x) == len(y):
        w, p = wilcoxon(x, y)
        print("Combination", key)
        print(p, w)
        print("____________________")
    else:
        print("not paired")
        print("____________________")
#significant combinations 
```
